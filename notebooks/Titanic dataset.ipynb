```python
def GetTitle(name):
    title_group = {'mr' : 'Mr',
                   'mrs' : 'mrs',
                   'miss' : 'Miss',
                   'master' : 'Master',
                   'don' : 'Sir',
                   'rev' : 'Sir',
                   'dr' : 'Offier',
                   'mme' : 'Mrs',
                   'ms' : 'Mrs',
                   'major': 'Offier',
                   'lady' : 'Lady',
                   'sir' : 'Sir',
                   'mlle' : 'Miss',
                   'col' : 'Offier',
                   'capt' :'Offier',
                   'the countess' : 'Lady',
                   'jonkheer' : 'Sir',
                   'dona' : 'Lady'
                   
                   }
    first_name = name.split(',')[1]
    title = first_name.split('.')[0]
    title = title.strip().lower()
    return title_group[title]
    
    
```

```python
df['Title'] = df.Name.map(lambda x : GetTitle(x))
```
